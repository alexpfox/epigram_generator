# load mongodb with quotes from 5 poets

```python
import pymongo
from pymongo import MongoClient
```

### import csvs/txts to store in mongo

```python
# texts/eliot_quotes.txt
# texts/frost_quotes.txt
# texts/kaur_quotes.csv
# texts/shakespeare_quotes.txt
# texts/wilde_quotes.csv
```

```python
import csv
import json
import sys

wilde_quotes = []
frost_quotes = []
kaur_quotes = []
eliot_quotes = []
shakespeare_quotes = []

with open('texts/wilde_quotes.csv', 'r') as csvfile:
    obj = csv.reader(csvfile, delimiter='\n')
    for row in obj:
        wilde_quotes.append('\n'.join(row))
        
with open('texts/frost_quotes.txt', 'r') as csvfile:
    obj = csv.reader(csvfile, delimiter='\n')
    for row in obj:
        frost_quotes.append('\n'.join(row))
        
with open('texts/kaur_quotes.csv', 'r') as csvfile:
    obj = csv.reader(csvfile, delimiter='\n')
    for row in obj:
        kaur_quotes.append('\n'.join(row))

with open('texts/eliot_quotes.txt', 'r') as csvfile:
    obj = csv.reader(csvfile, delimiter='\n')
    for row in obj:
        eliot_quotes.append('\n'.join(row))

with open('texts/shakespeare_quotes.txt', 'r') as csvfile:
    obj = csv.reader(csvfile, delimiter='\n')
    for row in obj:
        shakespeare_quotes.append('\n'.join(row))
```

```python
#make sure that worked
print(kaur_quotes)
```

### create and fill db

```python
# create db
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["corpora"]

# create collection for each poet
db.create_collection('wilde')
collection = db.get_collection('wilde')
collection.insert_many([{'quote': row} for row in wilde_quotes])

db.create_collection('frost')
collection = db.get_collection('frost')
collection.insert_many([{'quote': row} for row in frost_quotes])

db.create_collection('kaur')
collection = db.get_collection('kaur')
collection.insert_many([{'quote': row} for row in kaur_quotes])

db.create_collection('eliot')
collection = db.get_collection('eliot')
collection.insert_many([{'quote': row} for row in eliot_quotes])

db.create_collection('shakespeare')
collection = db.get_collection('shakespeare')
collection.insert_many([{'quote': row} for row in shakespeare_quotes])
```

```python
# check that it worked
db.list_collection_names()
```
